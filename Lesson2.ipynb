```python
import os
import csv
import pprint
```

```python
DATADIR = ""
DATAFILE = "beatles-diskography.csv"
```

```python
def parse_csv(datafile):
    data = []
    n = 0
    with open(datafile,'rb') as sd:
        r = csv.DictReader(sd)
        for line in r:
            data.append(line)
    return data
```

```python
datafile = os.path.join(DATADIR,DATAFILE)
pprint.pprint(parse_csv(datafile))
```
